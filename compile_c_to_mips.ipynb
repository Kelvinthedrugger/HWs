# Let's run!

## example: hello.c, one can neglect it
## just remember we use 'compile_mips(fname without .c)' to compile the code to mips assembly

```
%%writefile hello.c
#include<stdio.h>

int main(){
  printf("\nhello, world\n");
  return 0;
}
```

```
!ls
```

```
def compile_mips(fname):
  !mips-linux-gnu-gcc -O3 -S -mfp32 -march=R2000 $fname".c"
  !ls -l
  return
```

```
compile_mips("hello")
```

## Fib, unreasonably long

```
%%writefile fib.c
int fib(int n)
{
  if (n == 0) return 0;
  else if (n == 1) return 1;
  else return fib(n-1) + fib(n-2);
}
```

```
compile_mips("fib")
```
